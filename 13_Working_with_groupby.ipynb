```python
import pandas as pd
import numpy as np
```

```python
a = pd.DataFrame(np.arange(16).reshape(4,4),
                 index = [1,2,1,2],
                 columns=list('abcd'))
a
```

```python
b = a.groupby(a.index)
```

```python
b
```

```python
for i,j in b:
  print(i)
  print('=============')
  print(j)
```

```python
def f(x):
  return x.min()
b.agg(f)
```

```python
data=pd.DataFrame({"letter":list("ABC")*4,
                   "num":["one","two"]*6,
                   "d1":np.random.randn(12),
                   "d2":np.arange(10,33,2)})
```

```python
data
```

```python
group=data.groupby(["letter","num"])
```

```python
group_d1=group["d1"]
```

```python
group_d1.agg("mean")
```

```python
group_d1.agg(["mean","std",f])
```

```python
#naming the fucntions the way we want
group_d1.agg([("f_mean","mean"),
              ("f_std",np.std)])
```

```python
#applying desired functions to desiered columns
group.agg({"d1":["count","max","mean"],
           "d2":"sum"})

group.agg({'d1':["count","max","mean"],
            'd2':'sum'})
```

```python
data.groupby(["letter","num"],
             as_index=False).mean()
```

```python
math=pd.DataFrame({"Class":list("AB")*3,
                   "Stu":["Kim","Sam",
                          "Tim","Tom","John","Kate"],
                   "Score":[60,70,np.nan,
                            55,np.nan,80]})
math
```

```python
grp = math.groupby('Class')
```

```python
grp.mean()
```

```python
func = lambda f:f.fillna(f.mean())
```
