```python
s1 = pd.Series(np.arange(4),index = list('acde'))
s2 = pd.Series(np.arange(5),index = list('acefg'))
```

```python
s1,s2
```

```python
df1=pd.DataFrame(
    np.arange(6).reshape(2,3),
    columns=list("ABC"),
    index=["Tim","Tom"])
df2=pd.DataFrame(
    np.arange(9).reshape(3,3),
    columns=list("ACD"),
    index=["Tim","Kate","Tom"])
print(df1)
print(df2)
```

```python
df1+df2
```

```python
df1.A+df2.A
```

```python
#to replace the NaN values 
df1.add(df2,fill_value=0)
```
