```python
df = pd.read_csv('https://raw.githubusercontent.com/TirendazAcademy/PANDAS-TUTORIAL/main/DataSets/data4.txt')
df
```

```python
#we can skip the rows while reading the data
df = pd.read_csv('https://raw.githubusercontent.com/TirendazAcademy/PANDAS-TUTORIAL/main/DataSets/data4.txt',skiprows=[0,2])
df
```

```python
#usecolumns
df = pd.read_csv('https://raw.githubusercontent.com/TirendazAcademy/PANDAS-TUTORIAL/main/DataSets/data4.txt',sep=',',skiprows = [0,2], usecols=[1,0])
df
```

```python
#nrows: we can specify the number of rows we want to read
df = pd.read_csv('https://raw.githubusercontent.com/TirendazAcademy/PANDAS-TUTORIAL/main/DataSets/data4.txt',sep=',',skiprows = [0,2], usecols=[1,0] ,nrows=3)
df
```
