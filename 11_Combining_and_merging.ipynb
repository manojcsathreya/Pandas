```python
import pandas as pd
import numpy as np
```

```python
d1=pd.DataFrame(
    {"key":["a","b","c","c","d","e"],
     "num1":range(6)})
d2=pd.DataFrame(
    {"key":["b","c","e","f"],
     "num2":range(4)})
```

```python
print(d1,d2)
```

```python
pd.merge(d1,d2)
```

```python
pd.merge(d1,d2, on='key')
```

```python
d3=pd.DataFrame(
    {"key1":["a","b","c","c","d","e"],
     "num1":range(6)})
d4=pd.DataFrame(
    {"key2":["b","c","e","f"],
     "num2":range(4)})
```

```python
d3
```

```python
d4
```

```python
pd.merge(d3,d4,left_on='key1',right_on='key2')
```

```python
pd.merge(d1,d2,how='outer')
```

```python
pd.merge(d1,d2,how='left')
```

```python
pd.merge(d1,d2,how='right')
```

```python
df1=pd.DataFrame(
    {"key":["a","b","c","c","d","e"],
     "num1":range(6),
     "count":["one","three","two",
              "one","one","two"]})
df2=pd.DataFrame(
    {"key":["b","c","e","f"],
     "num2":range(4),
     "count":["one","two","two","two"]})
```

```python
pd.merge(df1,df2,on=['key','count'],how='outer')
```

```python
pd.merge(df1, df2, on="key", how='outer')
```

```python
pd.merge(df1, df2, 
         on='key', 
         suffixes=('_data1', '_data2'))
```

Merging on Index

```python
df1=pd.DataFrame(
    {"letter":["a","a","b",
               "b","a","c"],
     "num":range(6)}) 
df2=pd.DataFrame(
    {"value":[3,5,7]},
    index=["a","b","e"])
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,left_on='letter',right_index=True)
```

```python
right=pd.DataFrame(
    [[1,2],[3,4],[5,6]],
    index=["a","c","d"],
    columns=["Tom","Tim"])
left=pd.DataFrame(
    [[7,8],[9,10],[11,12],[13,14]],
    index=["a","b","e","f"],
    columns=["Sam","Kim"])
```

```python
right
```

```python
left
```

```python
pd.merge(right,left,right_index = True, left_index=True,how = 'outer' )
```

```python
left
```

```python
right
```

```python
left.join(right)
```
