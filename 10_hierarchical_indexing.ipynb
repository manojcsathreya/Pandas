```python
df=pd.DataFrame(
    np.arange(12).reshape(4,3),
    index=[["a","a","b","b"],
           [1,2,1,2]],
    columns=[["num","num","ver"],
             ["math","stat","geo"]])
df
```

```python
df.index.names = ['class','exam']
df.columns.names = ['field','subj']
```

```python
df
```

```python
df['num']
```

```python
df.swaplevel('class','exam')
```

```python
df
```

```python
df.sort_index(level=1)
```

```python
df.sum(level='exam')
```

```python
df.sum(level='field',axis=1)
```

```python
df.sum(level='subj',axis=1)
```

```python
data=pd.DataFrame(
    {"x":range(8),"y":range(8,0,-1),
     "a":["one","one","one","one","two",
          "two","two","two"],
     "b":[0,1,2,3,0,1,2,3]})
data
```

```python
#to used columns as index
data2 = data.set_index(['a','b'])
data2
```

```python
#if we also want the column to be preserved
data3 = data.set_index(['a','b'],drop=False)
data3
```

```python
#we can rever the changes of data2 as we don't a and b. We cannot reset the index of data3 as we have the columns named a and b
data2.reset_index()
```
